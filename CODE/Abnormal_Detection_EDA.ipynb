```python
import matplotlib.pyplot as plt
import seaborn as sns

test = pd.read_csv('/content/drive/MyDrive/공기압축기/test_data.csv')
test.drop(['type'],axis=1,inplace=True)
jh_box = pd.concat([test, predictions],axis=1)
#jh_box.drop(['type'],axis=1,inplace=True)
outliers = jh_box[jh_box['label'] == 1]
inliers = jh_box[jh_box['label'] == 0]

features = jh_box.columns[:-1]  # 'label' 열을 제외한 모든 피처들

for feature in features:
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=jh_box.index, y=jh_box[feature], hue=jh_box['label'], palette='Set1')
    plt.xlabel('Index')
    plt.ylabel(feature)
    plt.title(f'Scatterplot of {feature} by Anomaly Label')
    plt.show()

print("Outliers:")
print(outliers.describe())
print("\nInliers:")
print(inliers.describe())
```
